```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
phone_sales_df = pd.read_csv(r'C:\Users\user\Downloads\MyPortfolioProjects\Phones EDA Using Pandas\phones_data.csv')
phone_sales_df = phone_sales_df.sort_values(by = 'brand_name')
phone_sales_df
```

```python
#getting Samsung phones data
samsung_phones_df = phone_sales_df[phone_sales_df['brand_name'].str.contains('Samsung')]
samsung_phones_df = samsung_phones_df.filter(items=['brand_name','os','popularity','best_price','lowest_price','highest_price','memory_size','battery_size'])
samsung_phones_df
```

```python
#getting Apple phones data
apple_phones_df = phone_sales_df[phone_sales_df['brand_name'].str.contains('Apple')]
apple_phones_df = apple_phones_df.filter(items=['brand_name','os','popularity','best_price','lowest_price','highest_price','memory_size','battery_size'])
apple_phones_df
```

```python
apple_and_samsung_df = samsung_phones_df.merge(apple_phones_df, how='outer')
apple_and_samsung_df
```

```python
apple_and_samsung_df_dummy = apple_and_samsung_df.filter(items=['brand_name','highest_price','best_price','lowest_price'])
apple_and_samsung_df_dummy.set_index('brand_name')
apple_and_samsung_df_dummy = apple_and_samsung_df_dummy.groupby('brand_name')
apple_and_samsung_df_dummy.mean()
```

```python
apple_and_samsung_df_dummy_agg = apple_and_samsung_df_dummy.mean()
apple_and_samsung_df_dummy_agg.plot(kind='bar',title='Apple & Samsung Phones Prices', xlabel='Phone Brand', ylabel='Aggregate Price($)')
```

```python
apple_and_samsung_df_dummy2 = apple_and_samsung_df.filter(items=['brand_name','popularity'])
apple_and_samsung_df_dummy2.set_index('brand_name')
apple_and_samsung_df_dummy2 = apple_and_samsung_df_dummy2.groupby('brand_name')
apple_and_samsung_df_dummy_agg2 = apple_and_samsung_df_dummy2.mean()
apple_and_samsung_df_dummy_agg2.plot.barh(title='Popularity', ylabel='Phone Brand', xlabel='Average Popularity')
```
