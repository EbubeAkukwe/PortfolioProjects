```python
#replacing the dates with their respective month in mall_df dataframe
#date replace code
from datetime import datetime,date
x = mall_df
mall_df_copy = x
for i in mall_df_copy['Date'] :
    date1 = i
    date1 = date1.split('/')
    m = date1[0]
    d = date1[1]
    y = date1[2]
    if int(m) < 10 :
        m = '0' + m
    else :
        pass
    if int(d) < 10 :
        d = '0' + d
    else :
        pass
    date = y+'-'+m+'-'+d
    date = datetime.fromisoformat(date)
    month = datetime.strftime(date, "%B")
    mall_df_copy['Date'] = mall_df_copy['Date'].replace(i, month)
#mall_df_copy
```
